```python
from zipfile import ZipFile as zipfile
import librosa
import os
import soundfile as sf
import warnings
from tqdm import tqdm as tqdm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def write_wav(x, filename, sample_rate):
    if type(x) != np.ndarray:
        x = np.array(x)
    with warnings.catch_warnings():
        warnings.simplefilter("error")
        sf.write(filename, x, sample_rate, subtype = 'FLOAT', format = 'WAV')
```

```python
zip_folder = '../dataset_zip/'
eval_file_names = [file for file in os.listdir(zip_folder) if not file.find('evaluation.audio') == -1]
dev_file_names = [file for file in os.listdir(zip_folder) if not file.find('development.audio') == -1]
print(eval_file_names, dev_file_names)
```

```python
for filename in dev_file_names:
    filepath = os.path.join(zip_folder, filename)
    with zipfile(filepath, 'r') as zipper:
        listOfFileNames = zipper.namelist()
        for file in tqdm(listOfFileNames):
            zipper.extract(file, path = os.path.join('../dataset1/asc1/trainset1_temp'))
```

```python
eval_ex_filepath = '../dataset1/asc1/valset1_temp/TUT-acoustic-scenes-2016-evaluation/audio/'
dev_ex_filepath = '../dataset1/asc1/trainset1_temp/TUT-acoustic-scenes-2016-development/audio/'
```

```python
for filename in tqdm(dev_file_names):
    x, _ = librosa.core.load(os.path.join(dev_ex_filepath, filename), sr = 16000, mono = False)
    x = x.reshape((-1,2))
    write_wav(x, os.path.join('../dataset1/asc1/trainset1', filename), 16000)
```

```python
for filename in tqdm(eval_file_names):
    x, _ = librosa.core.load(os.path.join(eval_ex_filepath, filename), sr = 16000, mono = False)
    x = x.reshape((-1,2))
    write_wav(x, os.path.join('../dataset1/asc1/valset1', filename), 16000)
```
